```python
data4 = data3.pivot_table(index='Userid',columns='Book.Title',values='Rating')
```

```python
data4
```

```python
#Impute those NaNs with 0 values
data4.fillna(0,inplace=True)
data4.head()
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book_sim =1-pairwise_distances(data4.values,metric='cosine')
```

```python
book_sim
```

```python
#Store the results in a dataframe
book_sim_df = pd.DataFrame(book_sim)
```

```python
#Set the index and column names to user ids
book_sim_df.index = data3.Userid.unique()
book_sim_df.columns = data3.Userid.unique()
```

```python
book_sim_df.iloc[0:7,0:7]
```

```python
np.fill_diagonal(book_sim,0)
book_sim_df.iloc[0:7,0:7]
```

```python
# Most similar Users
book_sim_df.idxmax(axis=1)[0:7]
```
